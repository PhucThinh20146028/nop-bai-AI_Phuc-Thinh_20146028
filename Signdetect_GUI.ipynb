```python
import tkinter as tk
from tkinter import filedialog
from PIL import ImageTk, Image
import numpy as np
import tensorflow as tf
from keras.models import load_model

# Biển báo giao thông
classes = {
    1: 'Biển cấm đỗ xe',
    2: 'Biển cấm người đi bộ',
    3: 'Biển đường một chiều',
    4: 'Biển đường ưu tiên',
    5: 'Biển cấm bấm còi',
    6: 'Biển có công nhân làm đường',
    7: 'Biển cấm đi ngược chiều',
    8: 'Biển đỗ xe',
    9: 'Biển giao nhau tại bùng binh',
    10: 'Biển dừng',
}

class TrafficSignDetector:
    def __init__(self, model_path):
        self.model = load_model(model_path)
        self.window = tk.Tk()
        self.window.title("Traffic Sign Detector")
        self.window.geometry("400x400")
        
        self.create_widgets()
        
    def create_widgets(self):
        self.browse_button = tk.Button(self.window, text="Select image", command=self.browse_image)
        self.browse_button.pack(pady=10)
        
        self.image_label = tk.Label(self.window)
        self.image_label.pack()
        
        self.result_label = tk.Label(self.window, text="")
        self.result_label.pack(pady=10)
        
    def browse_image(self):
        file_path = filedialog.askopenfilename(filetypes=(("Image files", "*.jpg;*.jpeg;*.png"), ("All files", "*.*")))
        
        if file_path:
            img = Image.open(file_path).resize((300, 300))
            img = ImageTk.PhotoImage(img)
            self.image_label.configure(image=img)
            self.image_label.image = img
            
            predicted_class = self.detect_traffic_sign(file_path)
            self.result_label.configure(text="Predicted class: " + predicted_class)
        else:
            self.result_label.configure(text="No image selected")
            
    def detect_traffic_sign(self, image_path):
        img = Image.open(image_path).resize((30, 40))
        img = np.array(img) / 255.0
        img = np.expand_dims(img, axis=0)
        
        prediction = self.model.predict(img)
        predicted_class = np.argmax(prediction)
        
        return classes[predicted_class]
    
    def run(self):
        self.window.mainloop()

# Chạy ứng dụng
if __name__ == '__main__':
    app = TrafficSignDetector('AI.h5')
    app.run()
```
